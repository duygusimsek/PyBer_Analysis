```python
#import dependencies
import matplotlib.pyplot as plt
```

```python
#Set the x-axis to a list of strings for each month.
x_axis=["Jan","Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

#Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot
plt.plot(x_axis,y_axis)
```

```python
#create the plot with ax.plot()
fig,ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
#Create a plot and add label for the legend
plt.plot(x_axis, y_axis, marker=".", color="blue", linewidth=1, label="Boston")

#Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Set the y limit between 0 and 45
plt.ylim,(0,45)

#Create a title 
plt.title("PyBer Fare by Month")

#Add the legend
plt.legend()
```

```python
#Using the object-oriented approach, make the following changes to your line chart:
#Create a plot and add label for the legend
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker="^", color="green", linewidth=2, label="Boston")

#Create labels for the x and y axes
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
#Set the y limit between 0 and 45
ax.set_ylim(0,45)

#Create a title 
ax.set_title("PyBer Fare by Month")

#Add the legend
ax.legend()

#Add grid lines
ax.grid()

```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Object-oriented approach
#Create the plot with ax.plot()
fig,ax=plt.subplots()
ax.bar(x_axis,y_axis)
```

```python
#to create a horizontal barchart
fig,ax=plt.subplots()
ax.barh(x_axis,y_axis)
```

```python
fig,ax=plt.subplots()
ax.barh(x_axis, y_axis, color="turquoise", label="Chicago")

ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")

ax.invert_yaxis()
```

```python
  plt.plot(x_axis, y_axis,"o")
```

```python
plt.scatter(y_axis, x_axis, c="r", label="Chicago")
plt.legend()
plt.xlabel("Date")
plt.ylabel("Fare ($)")
plt.title("PyBer Fare by Month")
plt.gca().invert_yaxis()
plt.xlim(0,50)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
#multiply each data point in the y-axis by 3
y_axis_larger=[]
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#for you can use this method
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig,ax = plt.subplots()
#transparency with alpha
ax.scatter(y_axis, x_axis, s=[i*5 for i in y_axis],linewidths=2, c="skyblue" , alpha=0.20, edgecolors="black", label="Boston",)
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
ax.set_xlim(0,50)
ax.invert_yaxis()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
colors = ["slateblue", "magenta", "lightblue", "green", "darkseagreen", "greenyellow", "yellow", "orange", "violet", "lime", "tomato", "mistyrose"]
explode_values= (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
colors = ["turquoise", "magenta", "lightblue", "green", "darkorange", "pink", "yellow", "orange", "violet", "darkseagreen", "tomato", "mistyrose"]

explode_values= (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
#for figsize ax
fig,ax = plt.subplots(figsize=(8, 8)) 
ax.pie(y_axis,
       explode=explode_values,
       colors=colors,
       labels=x_axis,
       shadow=True,
       startangle=95,
       counterclock=False,
       labeldistance=1.1,
       autopct='%.1f%%')
plt.show()
```
